```python
import matplotlib.pyplot as plt
```

```python
# Labels and corresponding values in counter clockwise direction
data = {'Apple': 26,'Mango': 17,'Pineapple': 21,'Banana': 29,'Strawberry': 11}
# List of corresponding colors
colors = ['orange', 'lightgreen', 'lightblue','gold', 'cyan']
```

Define a variable to highlight a section of the pie chart by separating it from the rest. If you donâ€™t want to highlight any sections, set all of the values to 0:

```python
# Needed if we want to highlight a section
explode = (0, 0, 0, 0, 0)
```

```python
# Plot the pie chart
plt.pie(data.values(), explode=explode, labels=data.keys(),colors=colors, autopct='%1.1f%%', shadow=False,startangle=90)
# Aspect ratio of the pie chart, â€˜equalâ€™ indicates tht we
# want it to be a circle
plt.axis('equal')
plt.show()
```

```python
# Needed if we want to highlight a section
explode = (0, 0.2, 0, 0, 0)
```

```python
# Plot the pie chart
plt.pie(data.values(), explode=explode, labels=data.keys(),colors=colors, autopct='%1.1f%%', shadow=False,startangle=90)
# Aspect ratio of the pie chart, â€˜equalâ€™ indicates tht we
# want it to be a circle
plt.axis('equal')
plt.show()
```

A pie chart is constructed by dividing a circle into slices whose angular amplitudes are proportional to the frequency classes. The areas identified by the slices are proportional to the frequencies. To make the diagram clearer, the various slices are filled with different colors.
